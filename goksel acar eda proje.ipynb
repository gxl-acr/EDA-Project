```python

columns_to_drop = ['Engine Fuel Type', 'Number of Doors', 'Market Category']

df = df.drop(columns=columns_to_drop)
```

```python
print("Original column names:")
print(df.columns)

df.rename(columns={
    'Engine Fuel Type': 'Fuel_Type',
    'Engine HP': 'H_Power',
    'Engine Cylinders': 'Cylinders',
    'Transmission Type': 'Transmission',
    'Driven_Wheels': 'Drive_Type',
    'Market Category': 'Market_Category',
    'Vehicle Size': 'Size',
    'Vehicle Style': 'Style',
    'highway MPG': 'Highway_MPG',
    'city mpg': 'City_MPG',
    'Popularity': 'Popularity',
    'MSRP': 'List_Price'
}, inplace=True)

print("New column names:")
print(df.columns)
```

```python
from sklearn.impute import KNNImputer
columns_to_impute = ['H_Power', 'Cylinders']
df_subset = df[columns_to_impute]
imputer = KNNImputer(n_neighbors=5)
df_imputed = imputer.fit_transform(df_subset)
df_imputed = pd.DataFrame(df_imputed, columns=columns_to_impute)
df[columns_to_impute] = df_imputed
print(df.head())
```

```python
import statsmodels.api as sm

cleaned_df = df.dropna() 
cleaned_df = cleaned_df[(cleaned_df['List_Price'] > 0)]

X = cleaned_df[['Year', 'H_Power', 'Cylinders', 'Highway_MPG', 'City_MPG', 'Popularity']]
y = cleaned_df['List_Price']
try:
    X = cleaned_df[['H_Power', 'Cylinders', 'Highway_MPG', 'Popularity']]
    y = cleaned_df['List_Price']

    # Check for missing values
    if X.isnull().values.any() or y.isnull().values.any():
        print("There are missing values in the data. Please clean the data before proceeding.")
    else:
        # Add a constant term
        X = sm.add_constant(X)

        # Fit the model
        model = sm.OLS(y, X).fit()

        # Print the summary
        print(model.summary())
except KeyError as e:
    print(f"Error: {e}. Ensure that the column names are correct.")
except Exception as e:
    print(f"An error occurred: {e}")
```
